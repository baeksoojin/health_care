```python
import pandas as pd
data_10=pd.read_csv('21년10월.csv')
data_10
```

```python
list_age = []
def make_age(age):
    if age<20 and age>=10:
      kind_age = "10대"
    elif age<30 and age>=20:
      kind_age = "20대"
    elif age<40 and age>=30:
      kind_age = "30대"
    elif age<50 and age>=40:
      kind_age = "40대"
    elif age<60 and age>=50:
      kind_age = "50대"
    elif age<70 and age>=60:
      kind_age = "60대"
    else: kind_age = "70대 이상"
    return kind_age

for i in data_10['TEST_AGE']:
    list_age.append(make_age(i))
    
    

data_10["kind_age"]=pd.DataFrame(list_age)
data_10
```

```python
#person2[결정장애]의 정보 -> 나이 40 몸무게 90 키 170 성 M






class person:
  def __init__(self,a,w,h,s):
     self.age = a
     self.weight = w
     self.height = h
     self.sex = s
    #데이터 전처리 : 사용자의 w,h를 입력받고 bmi 지수식을 활용해서 사용자의 bmi를 알아냄.
     self.bmi = w/((h*h)/10000)
        
        #데이터 전처리 : 사용자의 나이를 기반으로 사용자의 나이대를 AGRDE_FLAG_NM 속성의 값들로 분류하는 작업을 해줌.
  def make_age(self):
    if self.age<20 and self.age>=10:
      self.kind_age = "10대"
    elif self.age<30 and self.age>=20:
      self.kind_age = "20대"
    elif self.age<40 and self.age>=30:
      self.kind_age = "30대"
    elif self.age<50 and self.age>=40:
      self.kind_age = "40대"
    elif self.age<60 and self.age>=50:
      self.kind_age = "50대"
    elif self.age<70 and self.age>=60:
      self.kind_age = "60대"
    else: self.kind_age = "70대 이상"

        #데이터 전처리 : BMI 수치를 활용해서 체중 등급을 6단계로 나눕니다. => 그 기준은 BMI기준표에 의함.
  def make_bmi_(self):
    if self.bmi<18.5:
      self.kind_bmi="저체중"
    elif (self.bmi>=18.5 and self.bmi<23):
      self.kind_bmi="정상"
    elif (self.bmi>=23 and self.bmi<25):
      self.kind_bmi="비만전단계비만"
    elif (self.bmi>=25 and self.bmi<30):
      self.kind_bmi="1단계비만"
    elif (self.bmi>=30 and self.bmi<40):
      self.kind_bmi="2단계비만"
    else:
       self.kind_bmi="3단계비만"

  
  def target_grade(self, t):
    self.target=t



person2 = person(40,90,170,'M')
person2.make_bmi_()
person2.make_age()
person2.target_grade("3등급")


print("person2에 대한 정보")
print(person2.age,person2.weight,person2.height,person2.sex)
print(person2.kind_bmi)
print(person2.kind_age)
print(person2.target)



```

```python
#연령대부터 필터링
is_age = data_10['kind_age'] == person2.kind_age
age=data_10[is_age]
age
```

```python

#target_grade가 목표치니까 체력처방데이터 data_10에서 등급이 목표치인애들인 것만 뽑아냄
is_target=age['CERT_GBN']==person2.target
target = age[is_target]
target
```

```python
#성별을 기준으로 전처리
is_sex=target['TEST_SEX']==person2.sex
target2 = target[is_sex]
target2
```

```python
# target의 데이터중 운동데이터값에서 가장 많이 등장한 운동을 뽑아서 단계별로 총 3개 출력해줌.

data_10=pd.read_csv('10월전처리데이터.csv')

mvm=data_10['PRES_NOTE'].to_list()

mvm2=[]
for i in mvm:
    mvm2.append(i.split(":"))
#print(mvm2)

#mvm2[0]
#['준비운동', '누워서 머리 위로 팔꿈치 펴기 ,손목 펴기/굽히기,팔꿈치 굽히기,앉아서 다리 밀기,앉아서 다리 펴기,앉았다 일어서기,서서 균형잡기,상지 루틴 스트레칭,하지 루틴 스트레칭1']

list_main=[]
list_pre=[]
list_end =[]

for i in range(0,len(mvm2)):
    for j in range(0,len(mvm2[i])):
        if j==1:
            list_pre.append(mvm2[i][j])
        if j==2:
            list_main.append(mvm2[i][j])
        if j==3:
            list_end.append(mvm2[i][j])
        



```

```python
list_pre_2=[]
for i in list_pre:
    if "/ 본운동" in i:
        list_pre_2.append(i.split("/ 본운동")[0])
        
list_pre_2

list_pre_3=[]
for i in list_pre_2:
    temp = i.split(',')
    for j in temp:
        list_pre_3.append(j)
    
list_pre_3
set_pre=list(set(list_pre_3))
#print(set_pre)
list_pre_4=[[0]*2 for i in range(len(set_pre))]


for i in range(0,len(set_pre)):
    list_pre_4[i][0] = list_pre_3.count(set_pre[i])
    list_pre_4[i][1] = list_pre_3[i]
    
# count_pre.sort(reverse=True)
# print(count_pre)
list_pre_4.sort(key = lambda a:a[0],reverse = True)
print(list_pre_4)
```

```python



#top3출력
print("준비운동 단계의 운동을 추천해드립니다\n")
for i in range(0,3):
    print(i+1,"순위 : ",list_pre_4[i][1])
```
