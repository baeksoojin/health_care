```python
#어플에서 사용자의 정보를 입력받고 되고싶은 국민체력등급을 선택해서 목표를 이룰 수 있도록 단계별 운동을 추천해주는 서비스를 구현.

class person:
  def __init__(self,a,w,h,s):
     self.age = a
     self.weight = w
     self.height = h
     self.sex = s
    #데이터 전처리 : 사용자의 w,h를 입력받고 bmi 지수식을 활용해서 사용자의 bmi를 알아냄.
     self.bmi = w/((h*h)/10000)
        
        #데이터 전처리 : 사용자의 나이를 기반으로 사용자의 나이대를 AGRDE_FLAG_NM 속성의 값들로 분류하는 작업을 해줌.
  def make_age(self):
    if self.age<20 and self.age>=10:
      self.kind_age = "10대"
    elif self.age<30 and self.age>=20:
      self.kind_age = "20대"
    elif self.age<40 and self.age>=30:
      self.kind_age = "30대"
    elif self.age<50 and self.age>=40:
      self.kind_age = "40대"
    elif self.age<60 and self.age>=50:
      self.kind_age = "50대"
    elif self.age<70 and self.age>=60:
      self.kind_age = "60대"
    else: self.kind_age = "70대 이상"

        #데이터 전처리 : BMI 수치를 활용해서 체중 등급을 6단계로 나눕니다. => 그 기준은 BMI기준표에 의함.
  def make_bmi_(self):
    if self.bmi<18.5:
      self.kind_bmi="저체중"
    elif (self.bmi>=18.5 and self.bmi<23):
      self.kind_bmi="정상"
    elif (self.bmi>=23 and self.bmi<25):
      self.kind_bmi="비만전단계비만"
    elif (self.bmi>=25 and self.bmi<30):
      self.kind_bmi="1단계비만"
    elif (self.bmi>=30 and self.bmi<40):
      self.kind_bmi="2단계비만"
    else:
       self.kind_bmi="3단계비만"

  def select(self,s1,s2,s3):
    self.step1 = s1
    self.step2 = s2
    self.step3 = s3
  
  def target_grade(self, t):
    self.target=t

  def solution(self,so1,so2,so3):
    self.solution1 = so1
    self.solution2 = so2
    self.solution3 = so3
  
  def print_data(self):
    print("\n당신의 운동방법을 알려드립니다.")
    print("준비운동단계 : "+self.solution1+"을 실천하세요")
    print("본운동단계 : "+self.solution2+"을 실천하세요")
    print("마무리운동단계 : "+self.solution3+"을 실천하세요")
    
#예시 => 특정 사람 person1이 가입시 나이,키,몸무게,성별을 입력한다.

person1 = person(18,80,160,'F')
person1.make_bmi_()
person1.make_age()
person1.target_grade('2등급')  #1등급, 2등급, 3등급, 참가증

# person1.bmi =>  bmi를 읽어옴// self.kind_bmi() => bmi단계를 불러옴.


person1.select(1,2,3) #step1이 준비운동에서의 단계값, 2,3가 그 다음단계를 의미
#step1,2,3가 사람들이 고른 단계를 의미함.


print("당신의 bmi 지수는 ", person1.kind_bmi, "이고,")
print("준비 운동 단계는 ", person1.step1, "단계,")
print("본운동 단계는 ", person1.step2, "단계,")
print("마무리 운동 단계는 ", person1.step3, "단계입니다.")

# person1의 나이가 10대이므로 10대 열만 뽑아냄
is_age = train_df['AGRDE_FLAG_NM'] == person1.kind_age
age=train_df[is_age]
age

# person1의 성별이 F이므로 F 열만 뽑아냄
is_sex = age['MBER_SEXDSTN_FLAG_CD'] == person1.sex
sex=age[is_sex]
sex

# person1의 bmi가 2단계비만이므로 2단계비만 열만 뽑아냄
is_bmi = sex['BMI_IDEX_GRAD_NM'] == person1.kind_bmi
bmi=sex[is_bmi]
bmi

#본인이 원하는 등급을 달성하기 위한 운동 선택
is_target=bmi['COAW_FLAG_NM'] == person1.target
target=bmi[is_target]
target

# FLAG_ACCTO_RECOMEND_MVM_RANK_CO 준비운동 본운동 마무리운동
# person1의 조건에 맞는 운동들만 출력 (준비운동)
is_step1 = target['FLAG_ACCTO_RECOMEND_MVM_RANK_CO'] == person1.step1   #
step1=target[is_step1]
is_step1_2 = step1['SPORTS_STEP_NM'] == "준비운동"
step1_2=step1[is_step1_2]
s1=step1_2['RECOMEND_MVM_NM']
#print(step1_2)

# person1의 조건에 맞는 운동들만 출력 (본운동)
is_step2 = target['FLAG_ACCTO_RECOMEND_MVM_RANK_CO'] == person1.step2   #
step2=target[is_step2]
is_step2_2 = step2['SPORTS_STEP_NM'] == "본운동"
step2_2=step2[is_step2_2]
s2=step2_2['RECOMEND_MVM_NM']
#print(step2_2)

# person1의 조건에 맞는 운동들만 출력 (마무리운동)
is_step3 = target['FLAG_ACCTO_RECOMEND_MVM_RANK_CO'] == person1.step3   #
step3=target[is_step3]
is_step3_2 = step3['SPORTS_STEP_NM'] == "마무리운동"
step3_2=step3[is_step3_2]
s3=step3_2['RECOMEND_MVM_NM']
#print(s3)

#최종적으로 person1이 어플에 입력한 몸무게,키,성별,나이를 기록하고 하고싶은 운동의 강도, 되고싶은 국민체력등급을 선택해 단계별로 운동을 안내해준다.
person1.solution(s1, s2, s3)
person1.print_data()

```
