# '운동명+링크'를 csv파일로 저장

```python
#운동명과 링크를 네이버 지식백과 운동처방동영상에서 크롤링해서 출력
import requests
from bs4 import BeautifulSoup as bs

mvm_list={}
name=[]  #운동명
site_link=[]  #사이트 링크

# keyword = keyword #person1.solution1
target_url = "https://terms.naver.com/list.naver?cid=58501&categoryId=58501&so=st4.asc"
# print(target_url)

data_page = requests.get(target_url)
soup3 = bs(data_page.content, "html.parser")

#print(soup3.select("#content div.contents_list_wrap>ul.contents_list>li.contents_sub>ul>li>ul>li>ul>li>a"))
for i in soup3.select("#content div.contents_list_wrap>ul.contents_list>li.contents_sub>ul>li>ul>li>ul>li>a"):
    print(i.get('href','/'))
    site_link.append(i.get('href', '/'))  #사이트 링크 출력
    print(i.get_text())
    name.append(i.get_text()) #운동명 출력
```

```python
# 크롤링한 정보를 csv 파일로 저장

import pandas as pd

mvm_list['name']=name
mvm_list['site_link'] = site_link
print(mvm_list.items())

mvmDF=pd.DataFrame(mvm_list)
mvmDF.to_csv("운동링크크롤링2.csv")
```

```python
#//*[@id="content"]/div[4]/ul/li[2]/ul/li[1]/ul/li[1]/ul/li[1]/a
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs

mvm_list={}
name=[]
site_link=[]

# keyword = keyword #person1.solution1
target_url = "https://terms.naver.com/list.naver?cid=58501&categoryId=58501&so=st4.asc"
# print(target_url)

data_page = requests.get(target_url)
soup3 = bs(data_page.content, "html.parser")

#print(soup3.select("#content div.contents_list_wrap>ul.contents_list>li.contents_sub>ul>li>ul>li>ul>li>a"))
for i in soup3.select("#content div.contents_list_wrap>ul.contents_list>li.contents_sub>ul>li>ul>li>ul>li>a"):
    #print(i.get('href','/'))
    site_link.append(i.get('href','/'))
    #print(i.get_text())
    name.append(i.get_text().strip())
    
mvm_list['name']=name
mvm_list['site_link'] = site_link
print(mvm_list.items())

mvmDF=pd.DataFrame(mvm_list)
mvmDF.to_csv("운동링크크롤링.csv")
```

```python
import urllib.request

info=[]
def crawl(url):
        soup=bs(urllib.request.urlopen(url).read(), 'html.parser')
        title = soup.find('h2', class_ = 'headword').get_text()
        soup2=soup.select("div.tmp_profile > table.tmp_profile_tb > tbody > tr > td")
        for i in soup2:
            info.append(i.get_text().strip())
        print(title)
 
# info 리스트에 저장된 값에서 단어들만 뽑아서 운동대상, 체력항목, 운동형태, 운동부위는 dictionary의 key 값으로,
# 나머지는 key에 해당하는 속성으로 저장하여 출력하기.

# URL = 'https://terms.naver.com/entry.naver?docId=3535305&categoryId=58501&cid=58501'  # 해당 운동의 정보 불러오기
# crawl(URL)

for i in site_link:
    crawl("https://terms.naver.com/"+i)

print("===============================================")
print(info)
    
```
